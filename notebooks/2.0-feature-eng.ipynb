```python
import sys, pathlib
ROOT_DIR = '../'
SRC_DIR = ROOT_DIR

sys.path.append(str(SRC_DIR))


from src.data_processing import build_feature_pipeline
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load data
df = pd.read_csv('../data/raw/transactions.csv')

# Run custom feature pipeline
pipe = build_feature_pipeline()
agg = pipe.fit_transform(df)

# Separate CustomerId for later use
customer_ids = agg['CustomerId']
X = agg.drop(columns=['CustomerId'])

# Scale numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df['CustomerId'] = customer_ids  # join back if needed

# Save
X_scaled_df.to_csv('../data/processed/customer_features.csv', index=False)
X_scaled_df.head()

```
